```python
import requests
import pandas as pd
import os
from dotenv import load_dotenv
load_dotenv(dotenv_path='/Users/monzon8/Documents/Ironhack/PROYECTOS/03_Proyecto/Proyect_LEA/.env')

api_key = os.getenv('api_keyTMDB')

# Función para obtener películas y series producidas por Marvel Studios
def get_marvel_studios_content():
    base_url = "https://api.themoviedb.org/3/discover/"
    
    # Parámetros para buscar películas y series de Marvel Studios
    params = {
        "api_key": api_key,
        "with_companies": 420,  # ID de Marvel Studios en TMDb
        "language": "en-US",
        "page": 1
    }
    
    # Almacenar los resultados
    results = []
    
    for media_type in ["movie"]:  # Para películas y series
        params["page"] = 1
        params["sort_by"] = "release_date.asc"  # Orden por fecha de estreno ascendente
        url = f"{base_url}{media_type}"
        
        while True:
            # Realiza la solicitud
            response = requests.get(url, params=params)
            data = response.json()
            
            # Recorre los resultados de la página actual
            for item in data["results"]:
                results.append({
                    "title": item.get("title") or item.get("name"),
                    "type": media_type,
                    "release_date": item.get("release_date"),
                    "overview": item.get("overview")
                })
            
            # Si no hay más páginas, termina el bucle
            if data["page"] >= data["total_pages"]:
                break
            else:
                params["page"] += 1

    # Crea un DataFrame con los datos recopilados
    dfmarvel = pd.DataFrame(results)
    return dfmarvel

# Ejecuta la función y guarda el resultado en un archivo CSV
dfmarvel = get_marvel_studios_content()
dfmarvel.to_csv("marvel_studios_movies.csv", index=False)
print("Archivo CSV guardado con éxito.")
```

```python
# Función para obtener películas y series relacionadas con Harry Potter
def get_harry_potter_content():
    base_url = "https://api.themoviedb.org/3/search/multi"
    query = "Harry Potter"
    
    # Parámetros para la solicitud
    params = {
        "api_key": api_key,
        "query": query,
        "language": "en-US",
        "page": 1,
        "include_adult": "false"
    }
    
    # Almacenar los resultados
    results = []
    
    while True:
        # Realiza la solicitud
        response = requests.get(base_url, params=params)
        data = response.json()
        
        # Recorre los resultados de la página actual
        for item in data["results"]:
            # Filtra solo películas
            if item["media_type"] in ["movie"]:
                results.append({
                    "title": item.get("title") or item.get("name"),
                    "type": item["media_type"],
                    "release_date": item.get("release_date"),
                    "overview": item.get("overview")
                })
        
        # Si no hay más páginas, termina el bucle
        if data["page"] >= data["total_pages"]:
            break
        else:
            # Pasa a la siguiente página
            params["page"] += 1

    # Crea un DataFrame con los datos recopilados
    dfHP = pd.DataFrame(results)
    return dfHP

# Ejecuta la función y guarda el resultado en un archivo CSV
dfHP = get_harry_potter_content()
dfHP.to_csv("harry_potter_movies.csv", index=False)
print("Archivo CSV guardado con éxito.")
```
