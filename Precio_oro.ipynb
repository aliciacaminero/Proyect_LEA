```python
import yfinance as yf
import pandas as pd
from datetime import datetime

# Fecha actual en formato YYYY-MM-DD
end_date = datetime.today().strftime('%Y-%m-%d')

# Descargar datos históricos del oro desde 1999 hasta hoy
gold_data = yf.download('GC=F', start='1999-01-01', end=end_date)
gold_data.to_csv('gold_price_1999_2024.csv')
```

```python
# Cargar el archivo con índice múltiple
gold_data = pd.read_csv('/Users/monzon8/Documents/Ironhack/PROYECTOS/03_Proyecto/Proyect_LEA/gold_price_1999_2024.csv', header=[0, 1], index_col=0)

# Seleccionar la columna 'Adj Close' con el ticker 'GC=F' y renombrar para simplificar
gold_data_adj_close = gold_data[('Adj Close', 'GC=F')].copy()
gold_data_adj_close.name = 'Adj Close'

# Convertir el índice a tipo datetime y extraer el año
gold_data_adj_close.index = pd.to_datetime(gold_data_adj_close.index)
gold_data_adj_close = gold_data_adj_close.to_frame()
gold_data_adj_close['Year'] = gold_data_adj_close.index.year

# Calcular la media anual de 'Adj Close'
annual_mean_price = gold_data_adj_close.groupby('Year')['Adj Close'].mean().reset_index()

# Renombrar columnas para mayor claridad
annual_mean_price.columns = ['Año', 'Media de Precio del Oro']

# Mostrar la tabla simplificada
annual_mean_price.head()
```

```python
# Calcular el crecimiento porcentual del valor del oro tomando el valor de 2024 como referencia

# Extraer el precio del oro en el año 2024 como referencia
precio_2024 = annual_mean_price.loc[annual_mean_price['Año'] == 2024, 'Media de Precio del Oro'].values

# Verificar si el valor de referencia existe para el año 2024 en el DataFrame
if len(precio_2024) == 1:
    precio_2024 = precio_2024[0]  # Asignar el valor escalar para el año 2024
    
    # Calcular el porcentaje de crecimiento en relación con el valor de 2024 para cada año
    annual_mean_price['Crecimiento % hasta 2024'] = ((precio_2024 - annual_mean_price['Media de Precio del Oro']) / annual_mean_price['Media de Precio del Oro']) * 100

    # Crear un nuevo DataFrame solo con las columnas solicitadas
    crecimiento_df = annual_mean_price[['Año', 'Crecimiento % hasta 2024']]
    

else:
    print("No se encontró un valor de referencia para el año 2024 en los datos.")
```

```python
import pandas as pd

# Cargar los archivos proporcionados
lego_data = pd.read_csv('/Users/monzon8/Documents/Ironhack/PROYECTOS/03_Proyecto/Proyect_LEA/Archivos de trabajo/top_10_HP.csv')
gold_growth = pd.read_csv('/Users/monzon8/Documents/Ironhack/PROYECTOS/03_Proyecto/Proyect_LEA/crecimiento_oro.csv')

# Mostrar las primeras filas de ambos archivos para verificar su estructura
lego_data.head(), gold_growth.head()
```

```python
# Renombrar columnas para estandarizar nombres y facilitar el análisis
lego_data = lego_data.rename(columns={'Año de Lanzamiento': 'Año', 'Porcentaje como nuevo': 'Revalorización LEGO'})
gold_growth = gold_growth.rename(columns={'Crecimiento % hasta 2024': 'Crecimiento Oro'})

# Fusionar los dos DataFrames en función del año
comparado_df = pd.merge(lego_data[['Año', 'Revalorización LEGO']], gold_growth[['Año', 'Crecimiento Oro']], on='Año', how='left')

# Evaluar si la revalorización de LEGO es mayor que el crecimiento del oro
comparado_df['Inversión LEGO Mejor'] = comparado_df['Revalorización LEGO'] > comparado_df['Crecimiento Oro']
```
